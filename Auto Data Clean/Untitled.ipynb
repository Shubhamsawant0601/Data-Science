```python
# Declaring the lists for columns names 
col_with_int_float_li = []
col_with_object_li = []


# Determing the type of column and sorting it into the respective list
for col_name, col_type in dict(df.dtypes).items():
    print(col_type)
    if "int" in str(col_type) or "float" in str(col_type):
        col_with_int_float_li.append(str(col_name))
    else:
        col_with_object_li.append(str(col_name))
        
        
df_with_int_and_float = df.drop(col_with_object_li, axis=1, inplace=False)
df_with_object = df.drop(col_with_int_float_li, axis=1, inplace=False)
```
